```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from scipy.optimize import minimize
import pdb
from __future__ import division
```

```python
test = sio.loadmat('test3.mat')
X = test['X']
y = test['y'].ravel()
Xval = test['Xval']
yval = test['yval'].ravel()
Xtest = test['Xtest']
ytest = test['ytest'].ravel()
```

```python
plt.plot(X.ravel(), y, 'r*')
plt.xlabel('Change in water level')
plt.ylabel('Water flowing out of the dam')
plt.show()
```

```python
def reg_cost(theta, X, y, lamb):
    m = len(y)
    X_ = np.concatenate((np.ones((X.shape[0], 1)), X), 1)
    temp1 = (X_.dot(theta[:, np.newaxis])-y[:, np.newaxis])
    return [1./2/m*np.sum(temp1**2) + lamb/2/m*np.sum(theta[1:]**2),
            1./m*X_.T.dot(temp1).ravel()
            + lamb/m*np.append(0, theta[1:])]
```

```python
reg_cost(np.array([1, 1]), X, y, 1.)
```

```python
res = minimize(reg_cost, np.array([0, 0]),         args=(
    X, y, 0), method='BFGS', jac=True,         options={'gtol': 1e-6, 'disp': True})
theta_ = res.x
```

```python
plt.plot(X.ravel(), y, 'r*')
plt.plot(X.ravel(), np.concatenate(
    (np.ones((X.shape[0], 1)), X), 1).dot(theta_[:, np.newaxis]).ravel())
plt.xlabel('Change in water level')
plt.ylabel('Water flowing out of the dam')
plt.show()
```

```python
train_error = np.zeros(y.shape)
val_error = np.zeros(y.shape)
for i in range(len(y)):
    res = minimize(reg_cost, np.array([0, 0]),
                   args=(X[:i+1, :], y[:i+1], 0), method='BFGS', jac=True,
                   options={'gtol': 1e-6, 'disp': False})
    theta_ = res.x
    train_error[i] = reg_cost(theta_, X[:i+1, :], y[:i+1], 0)[0]
    val_error[i] = reg_cost(theta_, Xval, yval, 0)[0]
plt.plot(np.arange(len(y)), train_error)
plt.plot(np.arange(len(y)), val_error)
plt.show()
```

```python
def poly_feat(X, p):
    poly=X.copy()
    for i in range(p-1):
        poly=np.concatenate((poly,X**(i+2)),1)
    return poly
```

```python
Xp = poly_feat(X, 8)
Xp = (Xp-np.mean(Xp, 0))/np.std(Xp, 0, ddof=1)
Xvalp = poly_feat(Xval, 8)
Xvalp = (Xvalp-np.mean(Xvalp, 0))/np.std(Xvalp, 0, ddof=1)
res = minimize(reg_cost, np.zeros((9,)),
               args=(Xp, y, 0), method='BFGS', jac=True,
               options={'gtol': 1e-6, 'disp': False})
theta_ = res.x
plt.plot(X.ravel(), y, 'r*')
temp = np.linspace(X.min(), X.max()+10, 100)
temp = poly_feat(temp[:, np.newaxis], 8)
temp = (temp-np.mean(temp, 0))/np.std(temp, 0)
plt.plot(np.linspace(X.min(), X.max()+10, 100),
         np.concatenate((np.ones((temp.shape[0], 1)), temp), 1).dot(theta_[:, np.newaxis]).ravel())
plt.xlabel('Change in water level')
plt.ylabel('Water flowing out of the dam')
plt.show()
```
