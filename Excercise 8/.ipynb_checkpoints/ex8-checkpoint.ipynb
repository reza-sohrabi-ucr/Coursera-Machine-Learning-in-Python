```python
def estimate_gaussian(X):
    return X.mean(0), X.var(0)


# def multivariate_gaussian(mu, var):
#     n=len(mu)
#     np.linalg.det(np.diag(var))
```

```python
from scipy.stats import multivariate_normal
mu, var = estimate_gaussian(X)
rv = multivariate_normal(mu, np.diag(var))
#x, y = np.mgrid[-1:1:.01, -1:1:.01]
x, y = np.mgrid[X[:, 0].min():X[:, 0].max():0.1,
                X[:, 1].min():X[:, 1].max():0.1]
u = x[:, 0]
v = y[0, :]
z = np.zeros((len(u), len(v)))
for i in range(0, len(u)):
    for j in range(0, len(v)):
        p = np.array([u[i], v[j]])
        # pdb.set_trace()
        z[i, j] = rv.pdf(p)
```

```python
plt.plot(X[:, 0], X[:, 1], 'bx')
plt.contour(x, y, z, 10)
plt.show()
```

```python
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import confusion_matrix


def select_threshold(Xval, yval, rv):
    pval = rv.pdf(Xval)
    maxf = 0
    for eps in np.linspace(pval.min(), pval.max(), 1000):
        y_pred = pval <= eps
        # pdb.set_trace()
        pr, re, fscore, _ = precision_recall_fscore_support(
            yval, y_pred, average='binary')
        #cf=confusion_matrix(yval, y_pred)
        if fscore >= maxf:
            maxf = fscore
            best_eps = eps
    return maxf, best_eps
```

```python
select_threshold(Xval, yval, rv)
```

```python
test = sio.loadmat('test13.mat')
Y = test['Y']
R = test['R']
test = sio.loadmat('test14.mat')
X = test['X']
Theta = test['Theta']
```

```python
def cofi_cost(params, Y, R, n=100, lamb=0):
    # first nm*n params are X and the rest are Theta
    nm, nu = R.shape
    # pdb.set_trace()
    X, Theta, _ = tuple(np.split(params, [n*nm, n*(nm+nu)]))
    X = X.reshape((nm, n))
    Theta = Theta.reshape((nu, n))
    cost = 1./2*(((X.dot(Theta.T)-Y)**2)*R).sum() + \\n        lamb/2.*((Theta**2).sum()+(X**2).sum())
    Jx = Theta.T.dot(((X.dot(Theta.T)-Y)*R).T).T+lamb*X
    Jt = X.T.dot(((X.dot(Theta.T)-Y)*R)).T+lamb*Theta
    return cost, np.append(Jx.flatten(), (Jt.flatten()))
```

```python
nm, nu, n = 1682, 943, 3
Xt = np.random.randn(nm, n)
Thetat = np.random.randn(nu, n)


def cost(par): return cofi_cost(par, Y, R, n, 10)


from scipy.optimize import minimize
res = minimize(cost, np.append(Xt.flatten(), (Thetat.flatten())),
               method='BFGS', jac=True,             options={'gtol': 1e-6, 'disp': False})
params = res.x
X, Theta, _ = tuple(np.split(params, [n*nm, n*(nm+nu)]))
X = X.reshape((nm, n))
Theta = Theta.reshape((nu, n))
```
