```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('ex2data1.txt', header=None)
data = (data.values)
X = data[:, 0:2]
X = np.concatenate((np.ones((len(X), 1)), X), 1)
Y = data[:, 2]
```

```python
posX = X[Y == 1, :]
negX = X[Y == 0, :]
plt.plot(posX[:, 1], posX[:, 2], 'r*')
plt.plot(negX[:, 1], negX[:, 2], 'b*')
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def cost(theta, X, Y):
    posX = X[Y == 1, :]
    negX = X[Y == 0, :]
    m = len(Y)
    return 1./m*(np.sum(-np.log(sigmoid(posX.dot(theta[:, np.newaxis])))) + sum(-np.log(1-sigmoid(negX.dot(theta[:, np.newaxis])))))


theta = np.array([0, 0, 0])


def grad(theta, X, Y):
    m = len(Y)
    return 1./m*np.dot(X.T, sigmoid(X.dot(theta[:, np.newaxis]))-Y[:, np.newaxis]).ravel()


grad(theta, X, Y)
```

```python
from scipy.optimize import minimize
x = minimize(cost, theta, args=(X, Y), method='BFGS', jac=grad,
             options={'gtol': 1e-6, 'disp': True})
```

```python
theta_ = x.x
X2b = -(theta_[1]*X[:, 1]+theta_[0])/theta_[2]
plt.plot(posX[:, 1], posX[:, 2], 'r*')
plt.plot(negX[:, 1], negX[:, 2], 'b*')
plt.plot(X[:, 1], X2b)
plt.show()
```

```python
data = pd.read_csv('ex2data2.txt', header=None)
data = (data.values)
X = data[:, 0:2]
degree = 6
out = np.ones((len(X), 1))
for i in range(1, degree+1):
    for j in range(0, i+1):
        temp = (X[:, 0]**(i-j))*(X[:, 1]**j)
        out = np.concatenate((out, temp[:, np.newaxis]), 1)
Y = data[:, 2]
```

```python
def cost_reg(theta, X,Y,lamb):
    posX=X[Y==1,:]
    negX=X[Y==0,:]
    m=len(Y)
    return 1./m*(np.sum(-np.log(sigmoid(posX.dot(theta[:,np.newaxis]))))\
                      +sum(-np.log(1-sigmoid(negX.dot(theta[:,np.newaxis])))))\
                      +lamb/2/m*np.sum(theta[1:]**2)

theta=np.zeros((out.shape[1],))
def grad_reg(theta, X,Y,lamb):
    m=len(Y)
    grad=1./m*np.dot(X.T,sigmoid(X.dot(theta[:,np.newaxis]))-Y[:,np.newaxis]).ravel()
    grad[1:]=grad[1:]+lamb/m*(theta[1:])
    return grad
x=minimize(cost_reg, theta, args=(out,Y,1.), method='BFGS', jac=grad_reg,\
                options={'gtol': 1e-6, 'disp': False})
theta_=x.x
pred=out.dot(theta_[:,np.newaxis])>=0
np.mean(pred.ravel()==Y)
```

```python
u = np.linspace(-1, 1.5, 50)
v = np.linspace(-1, 1.5, 50)


z = np.zeros((len(u), len(v)))

for i in range(0, len(u)):
    for j in range(0, len(v)):
        t = 1
        for k in range(1, degree+1):
            for l in range(0, k+1):
                temp = (u[i]**(k-l))*(v[j]**l)
                t = np.append(t, temp)
        z[i, j] = t.dot(theta_[:, np.newaxis])


z = z.T
```

```python
posX = X[Y == 1, :]
negX = X[Y == 0, :]
plt.plot(posX[:, 0], posX[:, 1], 'r*')
plt.plot(negX[:, 0], negX[:, 1], 'b*')
plt.contour(u, v, z, [0])
plt.legend(['pos', 'neg', 'bound'])
plt.show()
```
